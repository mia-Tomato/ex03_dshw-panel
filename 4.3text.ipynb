

## 4.3 核心结论

#### 1. 理论框架：A股上市公司支持优序融资理论，且产权性质显著调节该关系

基准模型（M1）及所有稳健性检验一致显示，盈利能力（NPR）的回归系数显著为负。这一方向性证据明确支持优序融资理论（Pecking Order Theory），即盈利能力越强的企业，其内源资金越充裕，从而降低对外部债务融资的需求。该结论在替换固定效应设定、加入宏观变量后依然稳健。

然而，这一关系因产权性质（SOE）而呈现显著异质性。交互项（NPR×SOE）系数在所有包含交互的模型中均显著为正（相对于民营企业），表明国有企业的杠杆率对盈利能力的敏感度显著低于民营企业。换言之，预算软约束、政策性负担或政府隐性担保可能削弱了国有企业内部资金对债务的替代作用，使其资本结构决策部分偏离优序融资理论的预测。分组回归的结果也与此一致：国有企业子样本中 NPR 负系数的绝对值更大，而交互效应进一步揭示了产权性质在边际上的调节方向。

#### 2. 时变特征：NPR‑Lev 关系在去杠杆与后疫情时期出现显著变化

时变系数模型（M4）揭示了盈利能力对杠杆率的边际影响并非一成不变。以2011年为基期，NPR的负向效应在2012‑2017年间相对稳定；然而在 **2018‑2019年** 出现了显著的正向偏离，即盈利‑杠杆之间的负相关关系明显减弱。这一时期恰与我国“去杠杆”政策深化、中美贸易摩擦加剧以及企业融资环境收紧重叠，表明在外部冲击下，企业被迫调整融资顺序，内源资金的缓冲作用受到抑制。

随后，在 **2023‑2024年**，NPR系数再次出现显著的负向偏离，意味着盈利‑杠杆负相关重新增强，甚至超过基期水平。这对应于后疫情时期货币政策的边际收紧、结构性去杠杆的继续推进，盈利能力再度成为企业降低债务依赖的重要来源。该时变规律说明，资本结构理论的有效性不能脱离宏观经济政策背景而孤立讨论。

#### 3. 计量方法：规模异质性存在，但对“小企业更优序”假设需谨慎解读

按企业规模中位数分组后，大规模企业的 NPR 负系数在绝对值上大于小规模企业，且两者均保持高度显著。这一结果本身没有直接证实“小企业更符合优序融资理论”的预期，反而提示大企业的杠杆决策对内部盈利的变动更为敏感。

对此的可能解释是：大规模企业融资渠道相对畅通，当盈利优异时，更容易主动替换为内源融资以降低财务风险；而小企业往往面临较强的融资约束，其杠杆水平更多取决于外部信贷可得性而非自身盈利状况，因此盈利对杠杆的调节作用相对较弱。门槛/分组分析表明，规模确实是影响盈利‑杠杆关系的重要因素，但其作用机制需要结合融资约束、企业生命周期等理论进一步探究。

#### 4. IFE 与 TWFE 比较：宏观货币环境是重要的遗漏变量

在基准模型中引入货币供应量增长率（M2_growth）后，NPR 系数的绝对值略有下降，且交互项的系数与显著性也发生了小幅变化。这一对比指明，如果不控制宏观货币环境，传统的双向固定效应模型可能会高估盈利能力对杠杆率的独立影响，因为宽松的货币时期会同时推高企业盈利和负债水平，产生偏误。

交互固定效应（IFE）通过吸收不可观测的宏观冲击，使得估计出的 NPR 系数更加干净地反映企业内部融资决策。因此，在资本结构研究中，考虑宏观经济变量不仅是稳健性要求，更是正确识别微观机制的必要条件。
